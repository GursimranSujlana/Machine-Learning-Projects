```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
```

```python
data_set = pd.read_csv('/Applications/Gursimran/Projects/ML-DL-AI-Hands-On-Projects/Finance/Stock-Sentimental-Analysis/data/data.csv', encoding = 'ISO-8859-1')
```

```python
#seeing whether the dataset is balanced or imbalanced
label_counts = data_set['Label'].value_counts()
print(label_counts)
```

```python
#dividing the training and testing datasets
training_set = data_set[data_set['Date'] < '20150101']
testing_set = data_set[data_set['Date'] > '20141231']
```

```python
# data preprocessing: removing punctuations or special characters
training_data = training_set.iloc[:,2:27]
training_data.replace("[^a-zA-Z]"," ", regex=True,inplace=True)
```

```python
# Combine headlines into one paragraph
headlines_train = []
for row in range(len(training_data.index)):
    headlines_train.append(' '.join(str(x) for x in training_data.iloc[row, 0:25]))
```

```python
# Implement CountVectorizer to convert sentences into vectors
countvector = CountVectorizer(ngram_range=(2, 2))
train_dataset = countvector.fit_transform(headlines_train)
```

```python
# Model training
randomclassifier = RandomForestClassifier(n_estimators=200, criterion='entropy')
randomclassifier.fit(train_dataset, training_set['Label'])
```

```python
# Data preprocessing for the testing set
testing_data = testing_set.iloc[:, 2:27]
testing_data.replace("[^a-zA-Z]", " ", regex=True, inplace=True)
```

```python
headlines_test = []
for row in range(len(testing_data.index)):
    headlines_test.append(' '.join(str(x) for x in testing_data.iloc[row, 0:25]))
```

```python
# Transform testing data
test_dataset = countvector.transform(headlines_test)
```

```python
test_dataset[0]
```

```python
predictions = randomclassifier.predict(test_dataset)
```

```python
predictions
```

```python
# Ensure the lengths of test['Label'] and predictions match
print("Length of test['Label']:", len(testing_set['Label']))
print("Length of predictions:", len(predictions))
```

```python
# Calculate confusion matrix, accuracy score, and classification report
matrix = confusion_matrix(testing_set['Label'], predictions)
print(matrix)
score = accuracy_score(testing_set['Label'], predictions)
print(score)
report = classification_report(testing_set['Label'], predictions)
print(report)
```
