```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the dataset
file_path = "../data/cleaned_water_quality.csv"
df=pd.read_csv(file_path)
```

```python
df.head()
```

```python
print(f"Loaded dataset shape: {df.shape}")
print(df.info())
```

```python
# summary statistics
print("Basic summary statistics")
df.describe(include="all")
```

```python
#count unique values
unique_counts = {
    "Sample Sites": df['Sample Site'].nunique(),
    "Analytes":df['Analyte'].nunique(),
    "Stream Group":df['Stream Group'].nunique(),
    "Units":df['Units'].nunique()
}
display(unique_counts)
```

```python
#distribution of analytes
plt.figure(figsize=(12,5))
df['Analyte'].value_counts().head(20).plot(kind="bar", color='c')
plt.title("Top 20 most common Analytes")
plt.xlabel("Analyte")
plt.ylabel("Count")
plt.show()
```

```python
#sample site distribution
plt.figure(figsize=(12,5))
df['Sample Site'].value_counts().head(20).plot(kind="bar", color='m')
plt.title("Top 20 Sample Sites with most measurements")
plt.xlabel("Sample Site")
plt.ylabel("Count")
plt.show()
```
